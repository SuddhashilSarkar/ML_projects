```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("/content/Salary_Data.csv")
print(data)
```

```python
sns.heatmap(data.isnull())
```

# **Data Cleaning**
Removing Null value and replacing NULL

```python
avg=data['Salary'].mean()
data['Salary'].fillna(avg,inplace=True)
```

```python
sns.heatmap(data.isnull())
```

```python
#.values converts to list
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
print(y)
```

```python
plt.scatter(x,y,color='orange')
plt.title("Salary Analysis")
plt.xlabel("Year of Exp")
plt.ylabel("Salary")
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(x,y)
```

```python
inp=[[2.5]]
predict_sal=model.predict(inp)
print(predict_sal)

inp1=[[7.5]]
predict_sal1=model.predict(inp1)
print(predict_sal1)
```

```python
plt.scatter(x,y,color='green')
plt.scatter(inp, predict_sal, color='red')
plt.scatter(inp1, predict_sal1, color='red')
plt.plot(x, model.predict(x), color='blue')
plt.title("Salary Analysis")
plt.xlabel("Year of Exp")
plt.ylabel("Salary")
```
